```python
#Importing the libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(color_codes=True)
```

```python
#importing the data
df = pd.read_csv("/content/Us_cars.csv")
df.head()
```

```python
#datatypes
df.dtypes
```

```python
#drop the irrelevant columns
df = df.drop('Vehicle Size',axis=1)
df.head()
```

```python
#Renaming the columns
df= df.rename(columns={"Engine HP":"HP", "Engine Cylinders":"Cylinders", "Transmission Type":"Transmission", "Driven_Wheels":"Drive Mode", "highway MPG":"MPG-H","city mpg":"MPG-C","MSRP":"Price"})
df.head()
```

```python
#shape
df.shape
```

```python
#Finding Duplicate rows
duplicate_rows=df[df.duplicated()]
duplicate_rows.shape
```

```python
#Counting the number of rows
df.count()
```

```python
# Drop duplicates rows
df = df.drop_duplicates()
df.head()
```

```python
#counting the total entries
df.count()
```

```python
# Finding Total number of empty rows
df.isnull().sum()
```

```python
# drop the missing value rows
df= df.dropna()
df.count(
)
```

```python
# finding the null value columns
df.isnull().sum()
```

```python
# Use Visualization Technique (Box Plot) to detect the outliers
sns.boxplot(x=df['MSRP'])
```

```python
sns.boxplot(x=df['Year'])
```

```python
#Manually finding the Interquartile range (IQR)
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
#Finding the relation between the data
df.corr()
```

##Heat Map

```python
#Adjusting the figure size
plt.figure(figsize=(10,5))
#Calculating the correlation
correlation = df.corr()
#displaying the correlation using heat map
sns.heatmap(correlation, cmap="BrBG",annot=True)
```

```python
#Data Visualization Bar Chart
df.Make.value_counts().nlargest(15).plot(kind="bar",figsize=(10,5))
plt.title("Price of the car by make")
plt.ylabel("Price of the car")
plt.xlabel("Make")
```

```python
#Pie Chart
df.Model.value_counts().nlargest(10).plot(kind="pie")
plt.show()
```
